# Classification

## Iris dataset

```python
from matplotlib import pyplot as plt
import numpy as np

from sklearn import datasets

#loading datasets
iris = datasets.load_iris()

print(iris.DESCR)
```

```python
features = iris.data
features
```

```python
labels = iris.target
labels
```

```python
#Training the classifier
from sklearn.neighbors import KNeighborsClassifier #KNN Classifier

clf = KNeighborsClassifier()

clf.fit(features, labels)

#Predicting with dataset
predicts = clf.predict(features)
print(predicts)
```

```python
from sklearn.metrics import mean_squared_error

print("The mean squared error is:", mean_squared_error(labels, predicts))
```

```python
#Plotting the graph

plt.scatter(features[:,1], labels)
plt.scatter(features[:,1], predicts)
plt.show()
```

## Using the CSV file and pandas:

```python
import pandas as pd
import seaborn as sns
sns.set_palette('husl')

iris = pd.read_csv("IRIS.csv")

g = sns.pairplot(iris, hue='species', markers='+')
plt.show()
```

```python
features = iris.iloc[:, :-1].values
labels = iris.iloc[:, -1].values
```

```python
from sklearn.neighbors import KNeighborsClassifier #KNN Classifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_iris = LabelEncoder()

labels = labelencoder_iris.fit_transform(labels)
onehotencoder = OneHotEncoder(categorical_features=[]) #Turning state name into a category

clf = KNeighborsClassifier()

clf.fit(features, labels)

#Predicting with dataset
predicts = clf.predict(features)
print("The mean squared error is:", mean_squared_error(labels, predicts))
```
