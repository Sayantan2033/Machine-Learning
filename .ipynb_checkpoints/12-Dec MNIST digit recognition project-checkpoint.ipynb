# MNIST dataset:

```python
#Fetching dataset
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784")
mnist
```

```python
#Creating features and labels from dataset
features = mnist.data
labels = mnist.target
```

```python
#plotting graph for a digit
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

digit = features[36001]

#Reshaping digit
digit = digit.reshape(28,28)
plt.imshow(digit, cmap=matplotlib.cm.binary, interpolation="nearest")

#Removing axis
plt.axis("off")
plt.show()

print(labels[36001])
```

```python
#Splitting data into training and testing sets

Features_Train, Features_Test = features[:60000], features[60000:]
Labels_Train, Labels_Test = labels[:60000], labels[60000:]

#Shuffling training set
import numpy as np

shuffle_index = np.random.permutation(60000)
Features_Train, Labels_Train = Features_Train[shuffle_index], Labels_Train[shuffle_index]
```

## Classifier:

```python
from sklearn.linear_model import LogisticRegression

LRC = LogisticRegression()

Labels_Train_2, Labels_Test_2 = (Labels_Train == "2"), (Labels_Test == "2")
LRC.fit(Features_Train, Labels_Train_2)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(LRC, Features_Train, Labels_Train_2, cv=2, scoring="accuracy")
```

```python
acc = np.mean([0.97673333, 0.97566667])
print("The accuracy is:", acc*100, "%")
```

```python
Labels_Predict = LRC.predict(Features_Test)
```

```python
#Question: Create a classifier that always classifies a digit as "not 2" with 90% "not 2"

two = 0
ntwo = 0
for i in Labels_Predict:
    if i:
        print("Is 2")
        two += 1
    else:
        print("Not 2")
        ntwo += 1
        
print("The percentage of not two is:", (ntwo/(ntwo+two))*100)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Labels_Test_2.astype(np.int8), Labels_Predict.astype(np.int8))
print("The MSE is:", mse)
```

```python
plt.scatter(range(len(Labels_Test_2[:50])), Labels_Test_2[:50], label="Actual value")
plt.scatter(range(len(Labels_Predict[:50])), Labels_Predict[:50], label="Predicted value")

plt.legend(loc="right")
plt.show()
```

## KNNC on MNIST:

```python
#Fetching dataset
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784")
mnist
```

```python
features = mnist.data
labels = mnist.target
```

```python
#Splitting data into training and testing sets

Features_Train, Features_Test = features[:60000], features[60000:]
Labels_Train, Labels_Test = labels[:60000], labels[60000:]
```

```python
#Training the classifier
from sklearn.neighbors import KNeighborsClassifier #KNN Classifier

KNNC = KNeighborsClassifier()

KNNC.fit(Features_Train, Labels_Train)

#Predicting with dataset
Labels_Predict = KNNC.predict(Features_Test)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Labels_Test, Labels_Predict)
print("The MSE is:", mse)
```
