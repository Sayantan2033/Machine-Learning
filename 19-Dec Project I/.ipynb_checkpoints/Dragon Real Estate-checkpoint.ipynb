```python
#Importing dataset
import pandas as pd

housing = pd.read_csv("Dragon Real Estate.csv")
housing.head()
```

```python
#Checking values in dataset
housing.shape
```

```python
#Checking for null values
housing.isnull().sum()
```

```python
#Filling null values
housing.RM.mean()
```

```python
housing[housing.RM.isnull()] = housing[housing.RM.isnull()].fillna(6.2878458498023795)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
housing.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#Preprocessing using sklearn and pandas

features = housing.iloc[:, :-1].values
labels = housing.iloc[:, -1:].values

from sklearn.model_selection import train_test_split

Features_Train, Features_Test, Labels_Train, Labels_Test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
#Finding Correlation
housing.corr()["MEDV"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix
attributes = ["MEDV", "RM", "ZN", "LSTAT"]
scatter_matrix(housing[attributes], figsize=(12,8))
plt.show()
```

```python
housing.plot(kind="scatter", x="RM", y="MEDV", alpha=0.8)
plt.show()
```

### Sklearn design:

    1. Estimator: Estimates some parameter based on a dataset. It has a fit method and transform method. Fit method fits the dataset and internally calculates parameters.
    2. Transformer: Transform method takes input and returns output based on learning techniques from fit function. It also has a convenient function called fit_transform() which fits and then transforms.
    3. Predictor: Model that predicts based on provided features. fit() and predict() are two common functions. There is also the score() function which will evaluate the model's predictions.

## Feature Scaling:

Two types:
1. Min-max scaling (normalizing) (value-min)/(max-min): sklearn provides a class called MinMaxScaler.
2. Standardization (value-mean): StandardScaler


```python
#Creating a pipeline

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(housing)
x = imputer.transform(housing)
housing_tr = pd.DataFrame(x,columns=housing.columns)

my_pipeline = Pipeline([('imputer', SimpleImputer(strategy="median")), ('std_scaler', StandardScaler())])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr.shape
```

## Models:

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

LRR = LinearRegression()
DTR = DecisionTreeRegressor(random_state=0)
RFR = RandomForestRegressor(random_state=0, n_estimators=100)

LRR.fit(Features_Train, Labels_Train)
DTR.fit(Features_Train, Labels_Train)
RFR.fit(Features_Train, Labels_Train)

Labels_LRR = LRR.predict(Features_Test)
Labels_DTR = DTR.predict(Features_Test)
Labels_RFR = RFR.predict(Features_Test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
print("The model scores are:\n")

print("Linear Regression:")
rmse = np.sqrt(mean_squared_error(Labels_LRR, Labels_Test))
r2 = r2_score(Labels_LRR, Labels_Test)
print("The root MSE is:",rmse)
print("The R^2 Score is:",r2)

print("\nDecision Tree Regression:")
rmse = np.sqrt(mean_squared_error(Labels_DTR, Labels_Test))
r2 = r2_score(Labels_DTR, Labels_Test)
print("The root MSE is:",rmse)
print("The R^2 Score is:",r2)

print("\nRandom Forest Regression:")
rmse = np.sqrt(mean_squared_error(Labels_RFR, Labels_Test))
r2 = r2_score(Labels_RFR, Labels_Test)
print("The root MSE is:",rmse)
print("The R^2 Score is:",r2)
```

## Saving the model:

```python
from joblib import dump, load
dump(RFR, "dragon.joblib")
```
